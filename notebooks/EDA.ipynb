| |corr| | Interpretation

```python
# Absolute correlation for strength
abs_corr = corr_target.abs().sort_values(ascending=False)


strong_threshold = 0.3
medium_threshold = 0.15

strong_predictors = abs_corr[abs_corr >= strong_threshold].index.tolist()
medium_predictors = abs_corr[(abs_corr < strong_threshold) & (abs_corr >= medium_threshold)].index.tolist()
weak_predictors = abs_corr[abs_corr < medium_threshold].index.tolist()

print("\nStrong predictors (|corr| >= 0.3):")
print(strong_predictors)

print("\nMedium predictors (0.15 <= |corr| < 0.3):")
print(medium_predictors)

print("\nWeak predictors (|corr| < 0.15):")
print(weak_predictors)
```

```python
from sklearn.feature_selection import mutual_info_classif

# Features and target
X = df.drop(columns=['Result'])
y = df['Result']

# Mutual information scores
mi_scores = mutual_info_classif(X, y, discrete_features=True, random_state=42)

mi_series = pd.Series(mi_scores, index=X.columns).sort_values(ascending=False)

print("\nMUTUAL INFORMATION SCORES:\n")
print(mi_series)

# Choose a threshold for "strong" MI
mi_threshold = mi_series.mean()   # or manually something like 0.05

strong_mi_predictors = mi_series[mi_series >= mi_threshold].index.tolist()

print("\nStrong predictors based on MI (>= mean MI):")
print(strong_mi_predictors)
```

```python
strong_corr_set = set(strong_predictors)
strong_mi_set = set(strong_mi_predictors)


# Intersection: features strong by all three methods
core_strong_features = strong_corr_set & strong_mi_set 

print("\nCore strong predictors (overlap of methods):")
print(core_strong_features)
```

```python


from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform
import matplotlib.pyplot as plt
```

```python
# Use only feature columns (exclude target)
X = df.drop(columns=["Result"])
feature_names = X.columns
```

```python
# Compute feature–feature correlation
corr = X.corr()

# Use absolute correlation (so -0.8 and +0.8 both mean strong relation)
corr_abs = corr.abs()

# Convert correlation to a distance matrix
# Distance = 1 - |correlation|
dist = 1 - corr_abs

# Convert the distance matrix to condensed form for linkage()
# squareform expects purely distance (diagonal 0, symmetric)
dist_condensed = squareform(dist.values, checks=False)
```

```python
# Perform hierarchical clustering
Z = linkage(dist_condensed, method="average")  # or "complete", "ward" (ward needs Euclidean)
```

```python
plt.figure(figsize=(10, 6))
dendrogram(
    Z,
    labels=feature_names,
    leaf_rotation=90
)
plt.title("Hierarchical Clustering of Features (Correlation-based)")
plt.tight_layout()
plt.show()
```

```python
# Choose a distance threshold for cutting the dendrogram
# Smaller threshold -> more clusters, larger -> fewer clusters
distance_threshold = 0.5   # you can experiment with this

cluster_labels = fcluster(Z, t=distance_threshold, criterion="distance")
```

```python
# Choose a fixed number of clusters
k = 3  # e.g. 3 clusters (URL structure, HTML behavior, domain trust)
cluster_labels = fcluster(Z, t=k, criterion="maxclust")
```

```python
# Map cluster labels to feature names
clusters = {}
for feature, label in zip(feature_names, cluster_labels):
    clusters.setdefault(label, []).append(feature)

# Print clusters
for label, feats in clusters.items():
    print(f"\nCluster {label}:")
    for f in feats:
        print("  -", f)
```
